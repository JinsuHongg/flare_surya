```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_file_path = "../data/surya-bench-flare-forecasting/train.csv"
df_train = pd.read_csv(train_file_path)

val_file_path = "../data/surya-bench-flare-forecasting/validation.csv"
df_val = pd.read_csv(val_file_path)

input_file_path = "../data/surya_input_data.csv"
df_input = pd.read_csv(input_file_path)
df_input = df_input.loc[df_input["present"] == 1]
```

```python
df_train = pd.merge(df_train, df_input, how="inner", left_on="timestamp", right_on="timestep")
df_val = pd.merge(df_val, df_input, how="inner", left_on="timestamp", right_on="timestep")
```

```python
print(f"number of training data {len(df_train)}")
print(f"number of validation data {len(df_val)}")
```
